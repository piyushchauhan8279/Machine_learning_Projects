```python
# Chance of Admission prediction:-
```

```python
# import library

import pandas as pd
```

```python
Admission=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
Admission.head()
```

```python
Admission.info()
```

```python
Admission.describe()
```

```python
# define y (target variable ) and x :-

y=Admission['Chance of Admit ']

```

```python
x=Admission.drop(['Chance of Admit ','Serial No'], axis=1)
```

```python
# split into training  and testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train
```

```python
y_train
```

```python
# model selection
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model=LinearRegression()
```

```python
# train the model
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# prediction
```

```python
y_prid=model.predict(x_test)
y_prid
```

```python
# accuracy
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_prid,y_test)
```

```python
# accuracy = 100%-4% = 96%
```
